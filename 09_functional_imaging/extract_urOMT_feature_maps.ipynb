# urOMT on a DICOM dataset — run, visualize, export

This notebook demonstrates the **urOMT** (unbalanced regularized Optimal Mass
Transport) workflow in pyCERR end to end:

1. **Load** a longitudinal DICOM series (the time points) + an ROI structure.
2. **Run** urOMT (Part 1 data prep → Part 2 solver → Eulerian / Lagrangian
   post-processing).
3. **Visualize** the Eulerian maps (speed, Péclet, density), the velocity
   vectors, and the Lagrangian pathlines inline.
4. **Export** the results to NIfTI — the scalar maps *and* the velocity vector
   field, aligned to the scan grid.

> Run this in the `pycerr` conda environment. For interactive 3-D, use the
> pyCERR viewer (`Tools → urOMT`, `Tools → 3D Visualization`).


## 1. Setup

Edit the paths below to point at your data.


```python
import os
import numpy as np
import matplotlib.pyplot as plt

import cerr.plan_container as pc
from cerr.uromt.config import buildConfig
from cerr.uromt.data import prepareData, scanTimeOrder
from cerr.uromt.solver import runUROMT
from cerr.uromt.analyze import runEULA, runEULAIntervals, runGLAD
from cerr.uromt import viz
from cerr.uromt.export import saveEulerianMapsNii, saveVectorFieldNii
from cerr.dataclasses.uromt import buildFromConfig, saveUROMTToPlan

# --- point these at your data ---------------------------------------------
DICOM_DIR = r"C:\software\urOMT\test_data\brain_pt\DICOM"
# ROI mask as a NIfTI (set to None if the DICOM already has an RTSTRUCT to use):
SEG_NII   = r"C:\software\urOMT\test_data\brain_pt\segmentation_mask_XC.nii"
OUT_DIR   = os.path.join(os.path.expanduser("~"), "uromt_out")
os.makedirs(OUT_DIR, exist_ok=True)
print("output ->", OUT_DIR)
```

## 2. Load the DICOM series + ROI

`loadDcmDir` reads every scan in the directory; for urOMT these are the
**longitudinal time points** (they must be co-registered onto one grid). The ROI
structure defines where transport is computed - here we load it from a NIfTI
mask, but any structure already in `planC` works.


```python
planC = pc.loadDcmDir(DICOM_DIR)
print("loaded %d scans (time points)" % len(planC.scan))

if SEG_NII:
    pc.loadNiiStructure(SEG_NII, 0, planC)        # add the ROI mask as a structure
structNum = len(planC.structure) - 1
print("ROI structure %d: %s" % (structNum, planC.structure[structNum].structureName))
```

## 3. Configure and run urOMT

The model/algorithm settings come from a JSON file
(`cerr/uromt/settings/uromt_model_settings.json`); the data settings (voxel
spacing in mm, ROI, time points) are read from `planC`. The default
normalization is **RSE** (relative signal enhancement, `S(t)/S0`) with the 1st
time point as the pre-contrast baseline.

Below we restrict the run to a few intervals for a quick demo via the `time`
selection (`first:jump:last`, 1-based into the time-ordered scans); drop that
line to transport every time point.


```python
scanNumV = scanTimeOrder(planC)                   # order scans by acquisition time
cfg = buildConfig(scanNumV, structNum)

# optional: a short window for a quick demo (transport time points 14,16,18,20)
cfg.settings["time"] = {"first_time": 14, "time_jump": 2, "last_time": 20}

cfg = prepareData(cfg, planC)                     # Part 1: crop, normalize, smooth
print("ROI grid:", cfg.trueSize,
      " spacing(mm):", [round(s, 3) for s in cfg.spacing],
      " frames:", len(cfg.vol))

res = runUROMT(cfg)                               # Part 2: the OMT solver
Eul = runEULAIntervals(res)                       # per-interval Eulerian maps
Lag = runGLAD(res)                                # Lagrangian pathlines
print("solved %d interval(s)" % len(res["u"]))

# (optional) store on planC so it survives a pickle and shows in the GUI
obj = buildFromConfig(cfg, res, runEULA(res), Lag)
runIdx = saveUROMTToPlan(planC, obj)
print("stored as planC.urOMT[%d]" % runIdx)
```

## 4. Visualize

### Eulerian maps
Per-interval, time-averaged maps on the ROI grid: advective speed `|v|`, the
Péclet number (advection vs diffusion) and the transported density.


```python
interval = 0
n = res["n"]
zc = n[2] // 2                                    # a middle slice

fig, axes = plt.subplots(1, 3, figsize=(13, 4))
for ax, (key, title, cmap) in zip(axes, [
        ("speed", "speed |v| (mm/t)", "turbo"),
        ("peclet", "Peclet (-)", "turbo"),
        ("rho", "density", "gray")]):
    m = np.ma.masked_equal(np.asarray(Eul[key][interval])[:, :, zc], 0.0)
    im = ax.imshow(m, cmap=cmap)
    ax.set_title(title)
    ax.axis("off")
    fig.colorbar(im, ax=ax, fraction=0.046, pad=0.04)
fig.suptitle("Eulerian maps - interval %d, slice %d" % (interval, zc))
plt.tight_layout()
plt.show()
```

### Velocity vectors
The time-mean OMT velocity for the interval, quivered over the density.


```python
u = res["u"][interval].mean(axis=2)               # (3, N) time-mean velocity
comp = [u[c].reshape(n, order="F") for c in range(3)]
U = comp[1][:, :, zc]                             # column (x) component
V = comp[0][:, :, zc]                             # row (y) component
mag = np.hypot(U, V)
s = max(1, min(n[0], n[1]) // 30)
yy, xx = np.mgrid[0:n[0], 0:n[1]]

fig, ax = plt.subplots(figsize=(6, 5))
ax.imshow(np.ma.masked_equal(np.asarray(Eul["rho"][interval])[:, :, zc], 0.0),
          cmap="gray")
q = ax.quiver(xx[::s, ::s], yy[::s, ::s], U[::s, ::s], V[::s, ::s],
              mag[::s, ::s], cmap="turbo", angles="xy", scale_units="xy",
              pivot="mid")
fig.colorbar(q, ax=ax, fraction=0.046, pad=0.04, label="|v| (mm/t)")
ax.set_title("velocity - interval %d, slice %d" % (interval, zc))
ax.axis("off")
plt.show()
```

### Lagrangian pathlines
Transport pathlines (effective velocity `v - sigma grad log rho`), projected in-plane and coloured by speed.


```python
fig = plt.figure(figsize=(6, 5))
viz.drawUROMTSlice(fig, res, Eul=runEULA(res), Lag=Lag, view="pathlines", axis=2)
plt.show()
```

## 5. Export to NIfTI

### Scalar Eulerian maps
One 3-D NIfTI per metric per interval (`<prefix>_<metric>_t<NN>.nii.gz`), placed
on the scan grid so they overlay the scan in any viewer.


```python
scanNum = cfg.frameScanNums[0]                    # any co-registered run frame
map_paths = saveEulerianMapsNii(Eul, planC, scanNum, OUT_DIR, prefix="uromt")
print("saved %d map file(s):" % len(map_paths))
for p in map_paths:
    print("  ", os.path.basename(p))
```

### Velocity vector field
The 3-component field saved as three scalar NIfTI files (row / column / slice
components) on the scan grid - load them together as a displacement/velocity
field in downstream tools.


```python
field = res["u"][interval].mean(axis=2)           # (3, N) time-mean velocity
vec_paths = saveVectorFieldNii(field, res["n"], res["bbox"], planC, scanNum,
                               OUT_DIR, prefix="uromt_velocity_t%02d" % (interval + 1))
print("saved velocity components:")
for p in vec_paths:
    print("  ", os.path.basename(p))
```
