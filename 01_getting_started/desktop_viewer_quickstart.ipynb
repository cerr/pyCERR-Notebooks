# pyCERR desktop viewer - quickstart

## Introduction

pyCERR ships an interactive **Qt desktop viewer** for browsing scans, structures and dose, contouring, fusion and 3-D viewing. It opens a native window, so it runs on your **local machine**, not on Colab / a headless server (for those, use the in-notebook viewers - see the last section).

> This notebook is a guide: the launch cell opens a desktop window and is meant to be run locally, not on Colab.

## Use the PyQt5 backend

pyCERR's desktop viewer uses **PyQt5**. Set this **before** importing the viewer (below) so the correct Qt binding is selected, especially if other Qt bindings are installed in your environment:

```python
import os
os.environ["QT_API"] = "pyqt5"
```

## Launch

**From a script** - `launch(planC)` blocks until you close the window and returns the (possibly edited) plan container:


```python
import cerr.plan_container as pc
from cerr import datasets
from cerr.viewer.pycerr_gui import launch

planC = pc.loadDcmDir(datasets.fetch_sample_data('head_and_neck'))
planC = launch(planC)          # interact, then close the window
print('structures now:', len(planC.structure))
```

**From IPython / Jupyter** - run `%gui qt` first, then `show(planC)` returns immediately and keeps a live handle while your session stays interactive:


```python
%gui qt
from cerr.viewer.pycerr_gui import show
viewer = show(planC)
# ... interact with the window ...
# viewer.planC          -> current plan container
# viewer.refresh_views()-> redraw after editing planC in code
```

## What you can do

* **Navigate** - scroll (or the on-screen arrows) to change slices; the layout shows axial / sagittal / coronal + a 3-D view.
* **Window/level** - presets (lung, bone, ...) or manual center/width.
* **Per-axis right-click menu** - choose the view orientation, scan, dose and structures shown in each panel (CERR's View/ScanSet/DoseSet/StructSet).
* **Structures panel** - toggle visibility, double-click to centre on a structure, and an optional **Current scan** filter.
* **Contouring** - freehand / polygon / brush draw and erase, with a live boundary and copy-from-adjacent-slice.
* **Fusion & overlays** - base-scan colormap/opacity and multi-scan overlays; **Registration QA** (mirror / side-by-side / alternate grid) under Tools.
* **Dose** - colorwash with an adjustable colorbar; **DVH** under Tools.
* **3-D view** - translucent orthogonal planes; double-click resets the camera.
* **Drag & drop import** - drop a DICOM folder, or a single RTSTRUCT file (imports only that structure), or NIfTI/.pkl.
* **File -> Export** - scan/dose/structures to NIfTI, structures to DICOM RTSTRUCT.
* **Planning** - IMRTP and ROE launchers under Tools.
