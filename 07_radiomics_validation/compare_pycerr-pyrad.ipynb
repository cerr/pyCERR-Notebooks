# Compare pyCERR and pyradiomics features

```python
import pandas as pd
import numpy as np
```

```python
cerrFeatFile = r'\\vpensmph\deasylab1\Aditya\HN_T2_radiomics\radiomics_features\wk0\ibsi_1_2.csv'
pyradFeatFile = r'\\vpensmph\deasylab1\Aditya\HN_T2_radiomics\radiomics_features\wk0\pyrad_feats.csv'

dfCerr = pd.read_csv(cerrFeatFile)
dfPyrad = pd.read_csv(pyradFeatFile)
```

```python
mappingFile = r'L:\Aditya\HN_T2_radiomics\radiomics_settings\pyradCerrMapping.json'

import json
 
# Opening JSON file
f = open(mappingFile)

pyradCerrFeatMap = json.load(f)
```

```python
import scipy
dfPyrad.sort_values(by=['id'], inplace=True)
dfCerr.sort_values(by=['id'], inplace=True)
corrList = []
featList = list(pyradCerrFeatMap.keys())
featsToCompare = []
pctDiffList = []
for key in featList:
    if pyradCerrFeatMap[key] != "":
        pyVals = dfPyrad[key].to_numpy()
        cerrVals = dfCerr[pyradCerrFeatMap[key]].to_numpy()
        if key == 'original_firstorder_Kurtosis':
            cerrVals += 3
        corrList.append(scipy.stats.pearsonr(pyVals, cerrVals)[0])
        pctDiff = np.mean(np.abs((pyVals - cerrVals)) / pyVals * 100)
        featsToCompare.append(key)
        pctDiffList.append(pctDiff)
    
```

```python
featsToCompare = np.array(featsToCompare)
pctDiffList = np.array(pctDiffList)
lowCorrV = np.array(corrList) < 0.96
lowCorrV = pctDiffList > 10
print(featsToCompare[lowCorrV])
print(pctDiffList[lowCorrV])
```
